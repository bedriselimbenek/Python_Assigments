```python
capital = int(input("Anaparayı giriniz: "))
btc = 19439.7  # 1 bitcoin = 19439.7$
profit = float(input("Kar giriniz: "))  # kar miktarını %7 ise 0.07 şeklinde giriniz.
day = int(input("Gün giriniz: "))
gain = btc * ((1 +profit) ** day) - btc
last_money = (btc + gain) + capital 
print("getiri: ", gain)
print(f"{day} gün sonundaki toplam para miktarı: {last_money}₺")
```
